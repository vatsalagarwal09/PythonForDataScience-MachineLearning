```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips, hue='sex', markers=['o','v'], scatter_kws={'s':50})
#scatter_kws helps to refer to scatter plot directly from seaborn
#Here we are changing size of markers through scatter_kws but we can do a large variety of things. For more information read documentation
```

```python
plt.figure(figsize=(40,25))
sns.lmplot(x='total_bill', y='tip', data = tips, col= 'day', row= 'time', hue='sex', aspect=0.5, height= 8)
```
