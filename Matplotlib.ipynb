```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)
plt.xlabel('XLabel')
plt.ylabel('YLabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
#(total numberOfRows, total numberOfColumns, plotNUmber)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#ObjectOrientedMethod
fig = plt.figure()#Using this command a blank canvas is created on which you can add further details through commands
axes = fig.add_axes([0.1,0.1,0.8,0.8])#Percentage of canvas to be used
#[Left, Bottom, Width, Height]
axes.plot(x,y)
axes.set_xlabel('XLabel')
axes.set_ylabel('YLabel')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('SquareFunction')
axes2.plot(y,x)
axes2.set_title('SquareRootFunction')
```

```python
#Object Oriented Method for subplots
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes can normally be used as an iterator or object
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()#Takes care of any overlapping between the plots
```

```python
fig = plt.figure(figsize=(5,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#Object Oriented Method for subplots
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()#Takes care of any overlapping between the plots
```

```python
fig.savefig('my_picture.png', dpi = 200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, label='XSquared')
axes.plot(x,x**3, label='XCubed')
axes.set_xlabel('XLabel')
axes.set_ylabel('YLabel')
axes.set_title('Set Title')

axes.legend(loc=10)
#loc is used to specify the location of legend box so that it does not overlaps with your curve
#You can use numbers from 0-10. Their specific locations are specified in legends documentation.
#Also you can use a (x,y) tuple to place the legend at exact position
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00', lw=0.5, alpha=0.75, ls=':', marker='o', markersize=20
        , markerfacecolor='purple', markeredgewidth=3, markeredgecolor='green')
#alpha is for transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00', lw=2, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
